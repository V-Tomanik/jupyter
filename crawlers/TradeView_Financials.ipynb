```python
import requests 
from bs4 import BeautifulSoup
```

```python
request = requests.get(url ='https://www.tradingview.com/symbols/BMFBOVESPA-ITUB3/',
                       headers= {'User-Agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:80.0) Gecko/20100101 Firefox/80.0'})

text = (request.text)
```

```python
sp = BeautifulSoup(text,'html.parser')
print(sp.prettify())
```

```python
print(sp.find_all('div',{'class':'tv-widget-fundamentals__item'}))
```
